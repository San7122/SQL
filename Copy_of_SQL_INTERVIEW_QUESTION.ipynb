**Basic SQL Questions**


**1 What is SQL?**


SQL (Structured Query Language) is a standard programming language used to manage and manipulate relational databases. It is used for querying, updating, and managing data, as well as controlling access to the database.

**2.Explain the difference between SQL and NoSQL.**

SQL: Uses structured query language for defining and manipulating data. It is used with relational databases, which store data in tables with predefined schemas. Examples include MySQL, PostgreSQL, and SQL Server.

NoSQL: Stands for "Not Only SQL" and refers to a variety of database systems that do not use SQL as their primary interface. These are often used for non-relational or distributed databases and can handle unstructured or semi-structured data. Examples include MongoDB, Cassandra, and Redis.


**3. What are the different types of SQL commands?**

DDL (Data Definition Language): Commands that define the structure of the database.

CREATE: Creates a new table, database, index, or view.

ALTER: Modifies an existing database object.

DROP: Deletes a database object.

DML (Data Manipulation Language): Commands that manipulate data stored in the database.

SELECT: Retrieves data from the database.

INSERT: Adds new data into a table.

UPDATE: Modifies existing data in a table.

DELETE: Removes data from a table.

DCL (Data Control Language): Commands that control access to the data.

GRANT: Gives user's access privileges to the database.

REVOKE: Removes access privileges given to the user.

TCL (Transaction Control Language): Commands that manage transactions in the database.

COMMIT: Saves all changes made in the current transaction.

ROLLBACK: Undoes changes made in the current transaction.

SAVEPOINT: Sets a savepoint within a transaction.

SET TRANSACTION: Sets the properties for a transaction.

**4.What is a database? What are tables and fields?**

Database: A structured collection of data stored electronically in a computer system, organized for efficient retrieval, insertion, and deletion.

Tables: Structures within a database that organize data into rows and columns. Each table represents a specific type of data.

Fields (Columns): Individual pieces of data stored in a table, representing the attributes of the data.

**5. What is a primary key? What are its characteristics?**
A primary key is a field (or combination of fields) that uniquely identifies each record in a table. Characteristics include:

Must contain unique values.

Cannot contain NULL values.

Each table can have only one primary key.

**6.What is a foreign key? How does it relate to a primary key?**

A foreign key is a field (or combination of fields) in one table that refers to the primary key in another table. It creates a relationship between the two tables, ensuring referential integrity.

**7. What is a unique key? How does it differ from a primary key?**

A unique key ensures that all values in a column are unique. The differences from a primary key include:

A table can have multiple unique keys.

Unique keys can contain NULL values, whereas primary keys cannot.

**8. Explain the concept of indexes. Why are they used?**

Indexes are database objects that improve the speed of data retrieval operations on a table. They work like an index in a book, allowing the database to find data more quickly without scanning the entire table. Indexes can be created on one or more columns of a table.

**9. What is a join in SQL? Explain different types of joins.**

Joins are used to combine rows from two or more tables based on related columns. Types of joins include:

INNER JOIN: Returns rows with matching values in both tables.

LEFT JOIN (LEFT OUTER JOIN): Returns all rows from the left table and the matched rows from the right table. Unmatched rows from the left table will have NULLs in columns of the right table.

RIGHT JOIN (RIGHT OUTER JOIN): Returns all rows from the right table and the matched rows from the left table. Unmatched rows from the right table will have NULLs in columns of the left table.

FULL JOIN (FULL OUTER JOIN): Returns all rows when there is a match in one of the tables. Rows with no match in the other table will have NULLs.

CROSS JOIN: Returns the Cartesian product of both tables, combining each row of the first table with every row of the second table.

**10.What is normalization? Explain the different normal forms.**

Normalization is the process of organizing data to reduce redundancy and improve data integrity. The normal forms include:

1NF (First Normal Form): Ensures that the table has a primary key and each column contains atomic (indivisible) values.

2NF (Second Normal Form): Meets 1NF criteria and ensures that all non-key columns are fully dependent on the primary key.

3NF (Third Normal Form): Meets 2NF criteria and ensures that all columns are dependent only on the primary key, eliminating transitive dependencies.

BCNF (Boyce-Codd Normal Form): A stricter version of 3NF where every determinant must be a candidate key.

4NF (Fourth Normal Form): Meets BCNF criteria and ensures no multi-valued dependencies.

5NF (Fifth Normal Form): Meets 4NF criteria and ensures no join dependencies.

**11.What is denormalization? Why and when is it used?**

Denormalization is the process of combining tables to reduce the number of joins and improve read performance. It is used when read performance is more critical than write performance, and when the database is subject to complex queries.

**12.What is a view in SQL? How is it different from a table?**
A view is a virtual table created by a query that combines data from one or more tables. Unlike a table, a view does not store data itself but provides a way to access data in tables. Views can be used to simplify complex queries, provide security by restricting access to specific data, and present data in a specific format.

**13. Explain the difference between DELETE and TRUNCATE commands **

DELETE: Removes specific rows from a table based on a condition. It can be rolled back and triggers are activated during its execution.

TRUNCATE: Removes all rows from a table. It cannot be rolled back, is faster than DELETE, and does not activate triggers.

**14.What is a stored procedure? How is it different from a function?**

A stored procedure is a set of SQL statements that can be executed as a single unit. Differences include:

Stored Procedure: Can perform actions (like modifying data) and return multiple values or none at all. It can also include control-of-flow statements.
Function: Must return a single value and cannot perform actions that modify data. It is used primarily for computations and cannot have side effects.

**15. What are transactions in SQL? Explain the ACID properties**.
Transactions are sequences of operations performed as a single logical unit of work. ACID properties include:

Atomicity: Ensures all operations within a transaction are completed successfully or none at all.

Consistency: Ensures the database remains in a consistent state before and after the transaction.

Isolation: Ensures that transactions are executed independently of each other.

Durability: Ensures that the results of a transaction are permanently saved in the database.

**20. Explain the use of the HAVING clause. How is it different from WHERE? **

FROM table_name
GROUP BY column1
HAVING COUNT(*) > 1;

**Advanced SQL Concepts**

**29 What are window functions in SQL? Provide examples.**

Window functions perform calculations across a set of table rows related to the current row. They are similar to aggregate functions but do not cause rows to become grouped into a single output row. Examples include ROW_NUMBER(), RANK(), and SUM() OVER (PARTITION BY ...). For instance:

SELECT employee_id, salary, RANK() OVER (ORDER BY salary DESC) AS rank
FROM employees;

**30. Explain the RANK(), DENSE_RANK(), and ROW_NUMBER() functions.**

RANK() assigns a rank to each row within a partition of a result set, with gaps in ranking if there are ties.

DENSE_RANK() assigns a rank without gaps; identical values receive the same rank.

ROW_NUMBER() assigns a unique number to each row, starting from 1 for the first row in each partition.

**31. What is a CTE (Common Table Expression)? How is it used?**
A CTE is a temporary result set defined within the execution scope of a SELECT, INSERT, UPDATE, or DELETE statement. It is defined using the WITH clause. Example:

WITH Sales_CTE AS (
  SELECT salesperson_id, SUM(sales_amount) AS TotalSales
  FROM sales
  GROUP BY salesperson_id
)
SELECT * FROM Sales_CTE WHERE TotalSales > 10000;

**32. Explain the difference between correlated and non-correlated subqueries.**

Correlated Subquery: A subquery that references columns from the outer query, executed once for each row of the outer query.

Non-Correlated Subquery: A subquery that can be run independently of the outer query and is executed only once.

**33. How do you handle NULL values in SQL?**

**34. What is a trigger in SQL? When would you use one?**

A trigger is a set of instructions automatically executed in response to certain events on a particular table or view, such as INSERT, UPDATE, or DELETE. Triggers are used for maintaining audit trails, enforcing business rules, and validating data.

**41. How would you write a query to calculate moving averages using SQL?**

Moving averages can be calculated using window functions.

 For example:

  AVG(sales) OVER (ORDER BY date ROWS BETWEEN 4 PRECEDING AND CURRENT ROW) AS MovingAvg

  SELECT salesperson_id, sales_amount,
    ROW_NUMBER() OVER (PARTITION BY region ORDER BY sales_amount DESC) AS rank
  FROM sales
)

FROM RankedSales
WHERE rank <= 3;

**51. How do you ensure data integrity in SQL databases?**

Data integrity is ensured through constraints (PRIMARY KEY, FOREIGN KEY, UNIQUE, CHECK), data types, and transactions. Using stored procedures and triggers also helps enforce business rules and maintain integrity.

**52. Explain the concept of data modeling and how it relates to SQL.**

**53. What are the differences between OLTP and OLAP databases? How does SQL usage differ in each?**

OLTP (Online Transaction Processing): Optimized for fast, efficient transactions and data integrity. SQL usage focuses on CRUD operations (Create, Read, Update, Delete).

OLAP (Online Analytical Processing): Optimized for complex queries and data analysis, often using a multi-dimensional schema like star or snowflake. SQL usage involves aggregation, complex joins, and large-scale data retrieval.

**54. How would you use SQL to generate a report summarizing key business metrics?**

Using aggregate functions and grouping data.


SELECT
  department,
  COUNT(*) AS total_employees,
  AVG(salary) AS average_salary,
  SUM(sales) AS total_sales
FROM employees
JOIN sales ON employees.employee_id = sales.employee_id
GROUP BY department;
